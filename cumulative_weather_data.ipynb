```
import pandas as pd
```

```
# 2019 data is downloaded from the Orchard House machine
src_jan = 'https://drive.google.com/uc?id=1Y1Ex6AVy_C-wzTYoLBpd3qjMVbs9ycZO'
src_jul = 'https://drive.google.com/uc?id=1M_vX51qflgskxMF7ZV3lOA2SLupooBiV'
# 2012 - 2018 weather data previously processed from Orchard House machine
src_prev = 'https://drive.google.com/uc?id=1_nB5yrKCNAXsjYJgjs9mE0Xd_x-OKCzJ'
```

```
# Load and concatenate data dumps from 2019
df_jan = pd.read_csv(src_jan, parse_dates=['Date Time '])
df_jul = pd.read_csv(src_jul, parse_dates=['Date Time '])
df_19 = df_jan.append(df_jul, ignore_index=True)
```

```
# Load existing weather archive from the Orchard House machine
df_prev = pd.read_csv(src_prev, parse_dates=['Date Time '])
```

## Combine
Concatenate weather archive with 2019

```
# select 2019 columns to match pre-existing stucture
existing_archive_columns = df_prev.columns
df_19 = df_19.loc[:, existing_archive_columns]
```

```
df_19
```

```
df = df_prev.set_index('Date Time ').loc[:'2018-12-31'].append(df_19.set_index('Date Time '))
```
